```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor

# Create synthetic dataset
rng = np.random.RandomState(42)
X = np.sort(5 * rng.rand(80, 1), axis=0)   # 80 random points between 0 and 5
y = np.sin(X).ravel()                      # target function (sin curve)
y[::5] += 0.5 - rng.rand(16)               # add some noise

# Fit Decision Tree Regressor
regressor = DecisionTreeRegressor(max_depth=3, random_state=42)
regressor.fit(X, y)

# Predict values
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]  # test input
y_pred = regressor.predict(X_test)

# Plot results
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="blue", label="Data (with noise)")
plt.plot(X_test, y_pred, color="red", linewidth=2, label="Decision Tree Prediction")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Decision Tree Regression Example")
plt.legend()
plt.show()
```
